```python
# define the node of the linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None
```

```python
class Deque:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
    
    #defines whether the queue is empty
    def isEmpty(self):
        return self.size <= 0
    
    #define the size of the queue
    def sizes(self):
        return self.size
    
    # enter the value from right hand side
    def pushLeft(self, data):
        self.size = self.size+1
        
        if not self.head:
            self.head = Node(data)
            self.tail = self.head
        else:
            newData = Node(data)
            self.head.prev = newData
            newData.next = self.head
            self.head = newData
    
    # enter the value from left hand side
    def pushRight(self, data):
        self.size = self.size+1
        
        if not self.tail:
            self.tail = Node(data)   
            self.head = self.tail
        else:
            newData = Node(data)
            self.tail.next = newData
            newData.prev = self.tail
            self.tail = self.tail.next
     
    # delete the most left value
    def popLeft(self):
        if self.isEmpty():
            return
        
        self.size = self.size-1
        retValue = self.head
        self.head = self.head.next
        
        if self.isEmpty(): # after seleting the value if there is no nodes than make the tail as None
            self.tail = None        
        
        return retValue.data
    
    # delete the most right value
    def popRight(self):
        if self.isEmpty():
            return
        
        self.size = self.size-1
        retValue = self.tail
        self.tail = self.tail.prev
        
        if self.isEmpty(): # after seleting the value if there is no nodes than make the head as None
            self.head = None
            
        return retValue.data        
```
